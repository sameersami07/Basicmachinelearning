#1. Basic Line Plot
o Plot a simple line graph of ùë¶ = ùë•
2
for ùë• values ranging from -10 to 10.
o Label the x-axis as "X Values" and the y-axis as "Squared Values."
o Add a title to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
x = np.arange(-10, 11)
y = x**2
plt.figure()
plt.plot(x, y, label='y = x^2')
plt.xlabel('X Values')
plt.ylabel('Squared Values')
plt.title('Line Plot of y = x^2')
plt.legend()
plt.grid()
plt.show()
```

#2. Bar Plot
o Create a bar chart showing the marks scored by 5 students: [Sai, 'Vamsi',
'Divij', 'Sritan', 'Sita'] with marks [85, 90, 78, 92, 88].
o Add labels to the x-axis and y-axis and give the chart a title.

```python
students = ['Sai', 'Vamsi', 'Divij', 'Sritan', 'Sita']
marks = [85, 90, 78, 92, 88]
plt.figure()
plt.bar(students, marks, color='skyblue')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks Scored by Students')
plt.show()
```

#3. Scatter Plot
o Generate two lists of 20 random integers each (for x and y coordinates).
o Create a scatter plot with these values, and customize it by adding color,
size, and labels to the points.


```python
x_scatter = np.random.randint(0, 100, 20)
y_scatter = np.random.randint(0, 100, 20)
sizes = np.random.randint(20, 200, 20)
colors = np.random.rand(20)
plt.figure()
plt.scatter(x_scatter, y_scatter, s=sizes, c=colors, alpha=0.5, cmap='viridis')
plt.colorbar(label='Color Intensity')
plt.xlabel('X Coordinates')
plt.ylabel('Y Coordinates')
plt.title('Random Scatter Plot')
plt.show()
```

#4. Histogram
o Simulate 100 random values from a normal distribution using
np.random.randn() and plot a histogram.
o Customize the number of bins and add a grid to the plot.

```python
data = np.random.randn(100)
plt.figure()
plt.hist(data, bins=15, color='purple', alpha=0.7)
plt.grid(True)
plt.title('Histogram of Random Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

#5. Pie Chart
o Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E']
with values [20, 30, 25, 15, 10].
o Display the percentage values on the chart and use a color palette for
differentiation.

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
plt.figure()
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Percentage Distribution of Categories')
plt.show()
```

#6. Box Plot (Seaborn)
o Create a box plot to visualize the distribution of the marks scored by
students in a class: [78, 85, 92, 88, 76, 95, 89, 84, 91, 73].
o Highlight the median and any outliers.

```python
marks_distribution = [78, 85, 92, 88, 76, 95, 89, 84, 91, 73]
plt.figure()
sns.boxplot(data=marks_distribution)
plt.title('Box Plot of Marks Distribution')
plt.show()
```

#7. Heatmap (Seaborn)

o Create a 5x5 matrix with random integers between 1 and 10 using
np.random.randint() and visualize it as a heatmap using Seaborn.

o Add a color bar and annotate the heatmap with the values.


```python
matrix = np.random.randint(1, 11, (5, 5))
plt.figure()
sns.heatmap(matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Random 5x5 Matrix')
plt.show()
```

# 8. Comparison of Matplotlib and Seaborn
o Generate a pair of plots (bar charts or line plots) using the same data in
both Matplotlib and Seaborn.
o Highlight the differences in the code, syntax, and aesthetics of the two
libraries.

```python
data = [10, 20, 15, 25, 30]
labels = ['A', 'B', 'C', 'D', 'E']

# Matplotlib Bar Plot
plt.figure()
plt.bar(labels, data, color='orange')
plt.title('Matplotlib Bar Plot')
plt.show()

# Seaborn Bar Plot
plt.figure()
sns.barplot(x=labels, y=data, palette='pastel')
plt.title('Seaborn Bar Plot')
plt.show()
```

#9.Pair Plot (Seaborn)
o Manually create a small dataset with columns such as 'Age', 'Height',
'Weight' for 10 individuals and visualize the relationships using a pair plot.


```python
import pandas as pd
data = {
    'Age': [25, 30, 35, 40, 22, 24, 29, 31, 28, 26],
    'Height': [5.5, 6.0, 5.8, 5.6, 5.4, 5.7, 5.9, 6.1, 5.8, 5.5],
    'Weight': [60, 72, 68, 75, 58, 62, 70, 80, 65, 64]
}
df = pd.DataFrame(data)
sns.pairplot(df)
plt.show()
```
